```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load the CSV file into a DataFrame
df_weather_kaggle = pd.read_csv('weatherHistory.csv')

# Display the first few rows of the DataFrame
df_weather_kaggle.head()
```

```python
# Display basic information about the DataFrame
df_weather_kaggle.info()


```

```python
# Display summary statistics of the numerical columns
df_weather_kaggle.describe()
```

```python
df_weather_kaggle.isnull().sum()
```

```python
# Plot a histogram
plt.hist(df_weather_kaggle['Temperature (C)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Temperature (C)')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a box plot
sns.boxplot(x='Precip Type', y='Temperature (C)', data=df_weather_kaggle)
plt.title('Box Plot of Temperature by Precip Type')
plt.show()
```

```python
# Create a scatter plot
plt.scatter(df_weather_kaggle['Humidity'], df_weather_kaggle['Wind Speed (km/h)'], c='coral', alpha=0.5)
plt.title('Scatter Plot of Humidity vs Wind Speed')
plt.xlabel('Humidity')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```

```python
# Set options to suppress FutureWarnings
pd.set_option('mode.chained_assignment', None)
pd.set_option('mode.use_inf_as_na', None)

# Select numeric columns
numeric_columns = df_weather_kaggle.select_dtypes(include='number')

# Correlation matrix heatmap with numeric columns
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
